```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset_inicial = pd.read_csv('services/covid_dataset_kaggle/country_wise_latest.csv')
```

```python
dataset_inicial
```

```python
dataset_mortes_regiao = dataset_inicial[['WHO Region','Deaths']]
```

```python
new_dataset = dataset_mortes_regiao.groupby('WHO Region',as_index = False).sum()
```

```python
y_range = new_dataset['Deaths'].max()
```

```python
new_dataset
```

```python

y_pos = np.arange(len(new_dataset['WHO Region']))

fig, ax = plt.subplots()
plt.bar(y_pos, new_dataset['Deaths'], align='center', alpha=1)
plt.xticks(y_pos, new_dataset['WHO Region'])
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')




plt.show()
```
